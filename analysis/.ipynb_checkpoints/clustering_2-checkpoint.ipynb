```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
data = pd.read_csv("/home/leon/Desktop/market/data/vector_data_ratings.csv", header=0, index_col=0)
```

```python
Z = linkage(data.values, 'ward')
```

```python
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=12.,  # font size for the x axis labels
)
plt.show()
```

```python
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=4,  # show only the last p merged clusters
    show_leaf_counts=True,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=12.,  # font size for the x axis labels
)
plt.show()
```

```python
last = Z[-10:, 2]
last_rev = last[::-1]
idxs = np.arange(1, len(last) + 1)
plt.plot(idxs, last_rev, label="distance")

acceleration = np.diff(last, 2)  # 2nd derivative of the distances
acceleration_rev = acceleration[::-1]
plt.plot(idxs[:-2] + 1, acceleration_rev, label="acceleration")
plt.legend()
plt.xlabel("number of clusters")
plt.ylabel("distance")
plt.show()

k = acceleration_rev.argmax() + 2  # if idx 0 is the max of this we want 2 clusters
print("clusters:{}".format(k))
```

```python
from scipy.cluster.hierarchy import fcluster
k = 4
cluster = fcluster(Z, k, criterion='maxclust')
```

```python
cluster = pd.DataFrame(cluster)
if data.shape[1] != 301:
    data = data.iloc[:,:-1]
data = pd.concat([data, cluster], axis=1)
```

# Calculating review distances from cluster

```python
cluster_data = []
for x in np.arange(1,k+1):
    cluster_data += [data[data.iloc[:,-1] == x]] 
```

```python
[(np.mean(x.iloc[:,0])/3, x.shape[0]) for x in cluster_data]
```

```python
cluster_max = np.array([np.amax(x, axis=0) for x in cluster_data]).astype(np.float32)
cluster_mean = np.array([np.mean(x, axis=0) for x in cluster_data]).astype(np.float32)
```

```python
for x in np.arange(k):
    cluster_max[x][0] = cluster_mean[x][0]
```

```python
from scipy.spatial.distance import cityblock, euclidean, cosine
```

```python
for x in np.arange(k):
    print(cluster_data[x].shape)
    dist = []
    cluster_squared = cluster_mean[x][:]**2
    for index, row in cluster_data[x].iterrows():
        dist.append(cosine(row.values[:]**2, cluster_squared))
    
    cluster_data[x]["distance"] = dist
        
        
        
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
figure(num=None, figsize=(8, 13), dpi=80, facecolor='w', edgecolor='k')
locat = 1
for reviews in cluster_reviews:
    plt.subplot(4, 2, locat)
    locat += 1
    ratings = [float(review["review_rating"]) for review in reviews]
    plt.hist(ratings, range = (1,5))
    if locat == 8:
        plt.xlabel("rating", fontsize=16)
    
    ratings = [float(review["review_rating"]) for review in reviews]

    lengths = []
    for rating in set(ratings):
        lengths += [[len(x["review_text"]) for x in reviews if float(x["review_rating"]) == rating]]
    
    maxi = np.max([x for y in lengths for x in y])
    plt.subplot(4, 2, locat)
    locat += 1
    plt.hist(lengths, range=(0,maxi*0.6), label=list(set(ratings)))
    plt.legend()
    if locat == 9:
        plt.xlabel("character length", fontsize=16)
        
plt.subplots_adjust(hspace = 0.3, wspace = 0.3)
plt.show()
```

# Extracting words of distances between clusters

```python
from numpy.linalg import norm
import spacy
nlp = spacy.load('en_core_web_md')

```

```python
def most_similar_vec(vector):
    word = nlp.vocab[u'dog']
    queries = [w for w in word.vocab if w.prob >= -14]
    sims = [cosine(vector, w.vector) for w in queries]
    similarity = [(x,y) for x, y in zip(queries, sims)]
    by_similarity = sorted(similarity, key=lambda x: x[1], reverse=False)
    return( [w[0].lower_ for w in by_similarity[:10]])
```

```python
connections = []
for vector1 in cluster_mean:
    connections.append([most_similar_vec((vector2 - vector1)[1:301]) for vector2 in cluster_mean])
    
```

```python
for x, word_list in zip(range(len(connections)), connections):
    df = pd.DataFrame([x[:6] for x in word_list])
    df.iloc[x,:] = ["-" for x in np.zeros(6)]
    print("\ncluster {} --------------------------------------------------------------".format(x))
    print(df)
```
